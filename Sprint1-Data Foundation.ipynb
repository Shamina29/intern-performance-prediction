```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\ssham\Downloads\intern_dataset_new (1).csv")
print("Shape:",df.shape)
print("Columns:",df.columns.tolist())
df.head()
```

```python
df.info()
```

```python
print("Missing Values")
print(df.isnull().sum())
```

```python
df['Completion_Time']=df['Completion_Time'].fillna(df['Completion_Time'].mean())
df['Feedback_Rating']=df['Feedback_Rating'].fillna(df['Feedback_Rating'].mean())
df['Attendance']=df['Attendance'].fillna(df['Attendance'].mean())
print("After Cleaning Missing Values")
print(df.isnull().sum())
print("All missing values handled")
```

```python
print("Duplicate")
print(df.duplicated().sum())
```

```python
print("Statistics")
print(df.describe())
```

```python
counts = df['performance_label'].value_counts()
print(counts)
colors={'High':'green','Medium':'orange','Low':'red'}
counts.plot(kind='bar',
            color=[colors[i] for i in counts.index],
            edgecolor='black')
plt.title("Intern Performance Distribution")
plt.xlabel("Performance Level")
plt.ylabel("Number of Interns")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
df.to_csv('intern_performance_cleaned.csv', index=False)
print("Dataset saved!")
```
